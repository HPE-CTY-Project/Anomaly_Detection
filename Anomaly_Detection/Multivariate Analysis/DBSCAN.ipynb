```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score

# Load the multivariate time series data
data = pd.read_csv('data (3).csv')
df=data
data=data.iloc[:,:(len(data.columns)-3)]
timestamp_col = data.pop('timestamp')


data=data.iloc[:,0::2]
print(data)
# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Define the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to the data
dbscan.fit(data_scaled)

# Extract the cluster labels and number of clusters
labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Print the number of clusters
print('Number of clusters:', n_clusters)

# Print the cluster labels

print('Cluster labels:', ', '.join(map(str, labels)))


```

```python
new_labels = [0 if label == -1 else 1 for label in labels]
print('New cluster labels:', '\n'.join(map(str, new_labels)))
```

```python
actual_labels=df.iloc[:,-1]
actual_labels
```

```python
accuracy = accuracy_score(actual_labels, new_labels)
precision = precision_score(actual_labels,new_labels)
recall = recall_score( new_labels,actual_labels)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
