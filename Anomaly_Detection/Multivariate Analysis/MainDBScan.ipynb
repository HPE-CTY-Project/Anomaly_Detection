```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
import time
import psutil
import os



```

```python
data = pd.read_csv('data (3).csv')
df=data
data=data.iloc[:,:(len(data.columns)-3)]
timestamp_col = data.pop('timestamp')


data=data.iloc[:,0::2]
print(data)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

start_wall_time = time.time()
start_cpu_time = time.process_time()

dbscan = DBSCAN(eps=0.7, min_samples=9)


dbscan.fit(data_scaled)


labels = dbscan.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)


print('Number of clusters:', n_clusters)



print('Cluster labels:', ', '.join(map(str, labels)),"\n\n")

new_labels = [1 if label == -1 else 0 for label in labels]
print('New cluster labels:', ' , '.join(map(str, new_labels)))

actual_labels=df.iloc[:,-3]



```

```python
accuracy = accuracy_score(actual_labels, new_labels)
precision = precision_score(actual_labels,new_labels)
recall = recall_score( new_labels,actual_labels)
f1=f1_score(new_labels,actual_labels)

```

```python
end_wall_time = time.time()
end_cpu_time = time.process_time()

```

```python
wall_time = end_wall_time - start_wall_time
cpu_time = end_cpu_time - start_cpu_time

```

```python
print(f'Physical Memory usage: {int(psutil.Process(os.getpid()).memory_info().rss / 1024 ** 2)} MB') 
print(f'Virtual Memory usage:  {int(psutil.Process(os.getpid()).memory_info().vms / 1024 ** 2)} MB')

count_anomalies = new_labels.count(0)
print('Number of anomalies:', count_anomalies)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1score",f1 )

print(f'Elapsed time: {int(wall_time)} seconds')
print(f'CPU time: {int(cpu_time)} seconds')
```
